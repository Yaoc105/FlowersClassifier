```python
for index in range(len(row)):
  name = flower_meanings_df.at[row[index], "Flower"]
  color = flower_meanings_df.at[row[index], "Color"]
  meaning = flower_meanings_df.at[row[index], "Meaning"]
  if color != "nan":
    print(f"If the {name} is {color}, then this flower's meaning is: {meaning}")
  else:
    print(f"In general, {name} have the flower's meaning of: {meaning}")
```
